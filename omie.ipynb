```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel('compras.xlsx')
df
```

```python
df = df.rename(columns={'Unnamed: 0':'Situação','Unnamed: 1':'Descrição','Unnamed: 2':'Família de Produto','Unnamed: 3':'Municipio','Unnamed: 6':'Preço Unitário de Venda','Unnamed: 7':'Estoque Disponível'})
df
```

```python
df.drop(columns=['Unnamed: 4','Unnamed: 5','Unnamed: 8','Unnamed: 9','Unnamed: 10'],index=[0,1],inplace=True)
df
```

```python
df.loc[6,'Municipio'] = 'Boca da Mata - AL'
df
```

```python
df['Preço Unitário de Venda'] = pd.to_numeric(df['Preço Unitário de Venda'])
df['Estoque Disponível'] = pd.to_numeric(df['Estoque Disponível'])
df
```

# Pergunta: Qual produto possui mais estoque?

```python
estoque = df.groupby(['Descrição']).sum()
estoque.sort_values('Estoque Disponível',ascending=False)
```

```python
plt.plot(estoque.sort_values('Estoque Disponível',ascending=False).drop(columns=['Preço Unitário de Venda'])[0:5],'or')
plt.plot(estoque.sort_values('Estoque Disponível',ascending=False).drop(columns=['Preço Unitário de Venda'])[0:5],'-b')
```

# Pergunta: Qual produto dentro do estoque é o mais caro?

```python
estoque = df.groupby(['Descrição']).sum()
estoque.sort_values('Preço Unitário de Venda',ascending=False)
```

```python
plt.plot(estoque.sort_values('Preço Unitário de Venda',ascending=False).drop(columns=['Estoque Disponível'])[0:3],'or')
plt.plot(estoque.sort_values('Preço Unitário de Venda',ascending=False).drop(columns=['Estoque Disponível'])[0:3],'-b')
```

# Pergunta: Qual familia possui mais estoque dentro do estoque?

```python
estoque = df.groupby(['Família de Produto']).sum()
estoque.sort_values('Estoque Disponível',ascending=False)
```

```python
plt.plot(estoque.sort_values('Preço Unitário de Venda',ascending=False).drop(columns=['Estoque Disponível'])[0:3],'or')
plt.plot(estoque.sort_values('Preço Unitário de Venda',ascending=False).drop(columns=['Estoque Disponível'])[0:3],'-b')
```
